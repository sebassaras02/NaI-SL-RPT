```python
#def find_val(x,y,con):
#    a=np.where(x>=con)
a1=[1,5,10,12]
a1=np.array(a1)
j=np.where(a1>=10)
def find_val(x,y,cond):
    if isinstance(x, np.ndarray) == True:
        pass
    else:
        x=np.array(x)
    if isinstance(y, np.ndarray) == True:
        pass
    else:
        y=np.array(y)
    j1=np.where(x>=cond)
    a=j1[0][0]
    return (y[a])
```
